##Задание 2
D&D - настольная ролевая игра. Большинство классов в D&D в какой-то момент получают возможность использовать заклинания, поэтому, если вы когда-либо будете играть в игру, вам придется просмотреть список заклинаний, которые ваш персонаж должен выучить. Данный датасет содержит в себе список всех заклинаний. Ниже будет приведен список колонок и описание каждой из них.

♦ Список всех колонок, содержащихся в датасете:

name - название заклинания

classes - какие классы могут выучить заклинание

level - уровень заклинания (0-й уровень означает, что это заговор)

school - какой школы магии это заклинание

cast_time - время накладывания заклинания

range - дальность заклинания

duration - длительность заклинания

verbal - Boolean, показывающий, что для сотворения заклинания требуется что-то произнести

somatic - Boolean, показывающий, что для сотворения заклинания требуется сделать жест

material - Bollean, показывающий, что для сотворения заклинания требуются материальные компоненты

material_cost - материальные компоненты, необходимые для создания

description - уникальное описание заклинания

Boolean - означает, что содержимает отображено в формате True со значением 1 или False со значением 0. Например, verbal 1, somatic 1 material 0 говорит, что для сотворения заклинания необходимо произнести волшебные слова, а также сделать особый жест.

##Бонус
У меня возникло желание сделать более красивую Heatmap на тему, сколько заклинаний определенных школ доступно каждому классу, а также Heatmap на тему, сколько заклинаний определенных уровней доступно каждому классу

```python
Spells = pd.read_csv("/content/drive/MyDrive/dnd-spells.csv",  lineterminator='\n')
Spells["class_arr"] = Spells.classes.map(lambda c:[cc.strip() for cc in c.split(",")])
class_spells = Spells.explode("class_arr")
class_spells.rename(columns={"class_arr":"class"}, inplace=True)
class_spells
```

```python
class_spells["count_me"] = 1
pivot = pd.pivot_table(class_spells[["class", "school", "count_me"]],
               values='count_me',
               index='class',
               columns='school',
               aggfunc="sum",
               fill_value = 0)
pivot
```

```python
plt.figure(figsize=(10,10))
sns.heatmap(pivot, annot = True )
```

```python
class_spells["count_me"] = 1
pivot = pd.pivot_table(class_spells[["class", "level", "count_me"]],
               values='count_me',
               index='class',
               columns='level',
               aggfunc="sum",
               fill_value = 0)
pivot
```

```python
plt.figure(figsize=(10,10))
sns.heatmap(pivot, annot = True )
```
