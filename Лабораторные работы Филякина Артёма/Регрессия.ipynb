2. Решение задачи регрессии на данных с использованием моделей sklearn (линейная регрессия + L1, L2), для моделей с регуляризациями подбор гиперпараметра. Вычисление значений метрик R^2, MAE, MSE, RMSE, MAPE. Выбор наилучшей модели.

```python
from sklearn.linear_model import LinearRegression, Lasso, Ridge
from sklearn.metrics import mean_squared_error, mean_absolute_error, mean_absolute_percentage_error
from math import sqrt
from sklearn.model_selection import GridSearchCV, RandomizedSearchCV
import numpy as np
```

```python
lr = LinearRegression().fit(X_train, y_train)
y_pred = lr.predict(X_test)
parameters = {'alpha': np.arange(0, 1, 0.1)}
lr.predict(X_test)
```

parameters - это словарь параметров, которые будут использоваться при настройке и обучении модели. В данном случае, это параметры, относящиеся к Ridge-регрессии, а именно - коэффициент регуляризации alpha. Они используются в методе GridSearchCV, который производит поиск по сетке параметров и выбирает те параметры, которые дают наилучшее качество модели на кросс-валидационной выборке. 

В данном случае, мы ищем наилучшие параметры модели Ridge-регрессии, чтобы улучшить качество предсказаний. Параметры задаются в виде словаря, где ключ - это название параметра, а значение - это список или массив значений, которые нужно проверить во время настройки модели. В данном случае мы задали параметр alpha как список значений от 0 до 1 с шагом 0.1.

Когда метод GridSearchCV запускается, он перебирает все возможные комбинации параметров и выбирает ту, которая дает наилучшее качество. В результате он возвращает объект ridge_optimal, у которого можно получить наилучшие параметры модели с помощью метода best_params_.

```python
ridge_optimal = GridSearchCV(Ridge(), parameters).fit(X_train, y_train)
ridge_optimal.best_params_
```

```python
#Ridge-L1
ridge = Ridge(alpha=0.9).fit(X_train, y_train)
y_pred = ridge.predict(X_test)
print(f'MAE: {mean_absolute_error(y_test, y_pred)}')
print(f'MSE: {mean_squared_error(y_test, y_pred)}')
print(f'RMSE: {sqrt(mean_squared_error(y_test, y_pred))}')
print(f'MAPE: {sqrt(mean_absolute_percentage_error(y_test, y_pred))}')
print(f'R^2: {ridge.score(X_test, y_test)}')
```

```python
#Lasso-L2
lasso = Lasso(alpha=0.4).fit(X_train, y_train)
y_pred = lasso.predict(X_test)
print(f'MAE: {mean_absolute_error(y_test, y_pred)}')
print(f'MSE: {mean_squared_error(y_test, y_pred)}')
print(f'RMSE: {sqrt(mean_squared_error(y_test, y_pred))}')
print(f'MAPE: {sqrt(mean_absolute_percentage_error(y_test, y_pred))}')
print(f'R^2: {lasso.score(X_test, y_test)}')
```

MAE (Mean Absolute Error) - средняя абсолютная ошибка. Это среднее арифметическое от абсолютных разностей между фактическими значениями и предсказанными значениями. Чем меньше MAE, тем лучше модель. 

MSE (Mean Squared Error) - средняя квадратичная ошибка. Это среднее арифметическое от квадратов разностей между фактическими значениями и предсказанными значениями. MSE измеряет среднюю ошибку квадрата между фактическими значениями и предсказанными значениями. Чем меньше MSE, тем лучше модель. 

RMSE (Root Mean Squared Error) - корень из средней квадратичной ошибки. RMSE является более интерпретируемой метрикой, чем MSE, поскольку она имеет ту же размерность, что и исходные данные, т.е. в случае регрессии - это единицы y. RMSE также измеряет среднюю ошибку между фактическими значениями и предсказанными значениями, но дает представление об их абсолютной ошибке. Чем меньше RMSE, тем лучше модель.

MAPE (Mean Absolute Percentage Error) - средняя абсолютная процентная ошибка. Это среднее арифметическое от абсолютных разностей между фактическими значениями и предсказанными значениями в процентном отношении к фактическим значениям. MAPE представляет собой процентную ошибку относительно фактических значений. Чем меньше MAPE, тем лучше модель.

R^2 (Coefficient of determination) - коэффициент детерминации. Он измеряет, насколько хорошо модель соответствует данным. Значение R^2 варьируется от 0 до 1, где 1 означает, что модель идеально соответствует данным. R^2 показывает отношение объясненной дисперсии к общей дисперсии и может быть использован для определения точности модели. Чем больше R^2, тем лучше модель.

3 - это степень многочлена, который будет использоваться для создания новых предикторов с помощью метода fit_transform объекта PolynomialFeatures. Эта степень определяет количество новых предикторов, которые будут созданы из каждого предиктора в исходном наборе данных. 

Например, если в исходном наборе данных есть только один предиктор, то после применения PolynomialFeatures(3) будут созданы три новых предиктора: первый - это исходный предиктор, второй - это его квадрат, а третий - его куб. Если исходный набор данных содержит несколько предикторов, то для каждого из них будет создано три новых предиктора, что может привести к значительному увеличению числа предикторов в итоговом наборе данных. 

Использование степеней многочлена различной величины позволяет создавать более сложные модели, которые могут более точно описывать зависимости между предикторами и выходными переменными. Однако при этом необходимо быть осторожным, чтобы не переобучить модель и избежать появления шума в данных.

```python
class LinearRegression() :
      
    def __init__( self, learning_rate, iterations, R_penalty) :
        self.learning_rate = learning_rate
        self.iterations = iterations
        self.R_penalty = R_penalty
              
    def fit( self, X, Y ) :
        
        # no_of_training_examples, no_of_features
          
        self.m, self.n = X.shape
        # weight initialization
          
        self.W = np.zeros( self.n )
        self.X = X
        self.Y = Y
          
          
        # gradient descent learning
                  
        for i in range( self.iterations ) :
            self.update_weights()
        return self
      
    # Helper function to update weights in gradient descent
    def predict( self, X ) :
      
        return X.dot( self.W )
    
    def update_weights( self ) :
             
        Y_pred = self.predict( self.X )
        error = self.Y - Y_pred
          
        # calculate gradients  
      
        dW = ( 2 * ( self.X.T ).dot( error )  ) / self.m + ( 2 * self.R_penalty * self.W ) / self.m  
          
        # update weights
      
        self.W = self.W - self.learning_rate * dW
          
        return self
    def metriks(self, y_p, y_tr) :
        print("MAE: ", abs(y_tr-y_p)/y_p.shape[0])
        print("MSE: ", ((y_tr-y_p)**2)/y_p.shape[0])
        print("RMSE: ", sqrt((y_tr-y_p)**2)/y_p.shape[0])
        print("MARE: ")
```
