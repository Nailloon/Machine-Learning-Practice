```python
import pandas as pd
import numpy as np
```

```python
map_dict = {
    False: 0,
    True: 1
}
for y in data1.columns:
    if(data1[y].dtype == 'bool'):
        data1[y] = data1[y].map(map_dict)
    if(data1[y].dtype == 'object'):
        data1 = pd.get_dummies(columns=[y], data = data1)
data1
```

##Первый пункт

```python
y = data1["price_usd"]
X = data1.drop(["price_usd"], axis=1)
```

```python
y
```

```python
X
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.5)
```

##Второй пункт

```python

from sklearn.linear_model import LinearRegression, Lasso, Ridge
from sklearn.metrics import mean_squared_error, mean_absolute_error, mean_absolute_percentage_error
from math import sqrt
lr = LinearRegression().fit(X_train, y_train)
lr.predict(X_test)
```

```python
y_pred = lr.predict(X_test)
print(f'MAE: {mean_absolute_error(y_test, y_pred)}') #средняя абсолютная ошибка
print(f'MSE: {mean_squared_error(y_test, y_pred)}') #средняя квадратичная ошибка
print(f'RMSE: {sqrt(mean_squared_error(y_test, y_pred))}') #квадратный корень из средней квадратичной ошибки
print(f'MAPE: {sqrt(mean_absolute_percentage_error(y_test, y_pred))}') #средняя абсолютная ошибка в процентах
print(f'R^2: {lr.score(X_test, y_test)}') #коэффициент детерминации
```

```python
#Собственная реализация
class LinReg:

    def __init__(self, lr=0.001, iters=5000, eps = 0.01):
        self.lr = lr
        self.iters = iters
        self.eps = eps

    def fit(self, x, y):
        dist = np.inf
        X = x.values
        self.X = X
        w = np.zeros(X.shape[1])
        iter = 0
        while dist > self.eps and iter <= self.iters:
            loss = self.fun_error(X, y, w)
            w = w + (self.lr * 2 * np.dot(X.T, np.dot(X, w) - y) / X.shape[0])
            dist = np.abs(loss - self.fun_error(X, y, w))
            iter += 1
        self.w = w

    def fun_error(self, x, y, w):
        return sum((y - np.dot(x, w)) ** 2) / x.shape[0]

    def predict(self, x):
        return np.dot(x, self.w)
```

```python
class LinRegr:

    def __init__(self, lr = 0.000001, n_iters=100):
        self.lr = lr
        self.n_iters = n_iters
        self.weights = None

    def fit(self, X, y):
        n_samples, n_features = X.shape
        self.weights = np.zeros(n_features)

        for _ in range(self.n_iters):
            y_pred = np.dot(X, self.weights)
            dw = (1/n_samples) * np.dot(X.T, (y_pred-y))

            self.weights = self.weights - self.lr * dw

    def predict(self, X):
        y_pred = np.dot(X, self.weights)
        return y_pred
```
